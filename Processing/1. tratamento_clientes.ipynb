```python
import pandas as pd
import numpy as np
```

```python
vendas = pd.read_csv('/Volumes/databricks_database/volumes/arquivos/Vendas.csv')
clientes = pd.read_csv('/Volumes/databricks_database/volumes/arquivos/Clientes.csv')
avaliacoes = pd.read_csv('/Volumes/databricks_database/volumes/arquivos/Avaliacoes.csv')
devolucoes = pd.read_csv('/Volumes/databricks_database/volumes/arquivos/Devolucoes.csv')
```

```python
clientes.info()
clientes.head(10)
```

```python
vendas.info()
vendas.head(10)
```

```python
vendas['Valor_Total'] = vendas['Quantidade'] * vendas ['Preço Unitário']

vendas_agg = vendas.groupby('Cliente_ID').agg(
    total_compras=('Venda_ID','count'),
    total_gasto=('Valor_Total','sum'),
    qtd_total_itens=('Quantidade','sum')
).reset_index()
```

```python
vendas_agg.info()
```

```python
devolucoes_merge = devolucoes.merge(
    vendas[['Venda_ID', 'Cliente_ID']],
    on='Venda_ID',
    how='left'
)
devolucoes_merge = devolucoes_merge.rename(
    columns={'Cliente_ID_y': 'Cliente_ID'}
)
```

```python
devolucoes_merge.columns
```

```python
devolucoes_agg = devolucoes_merge.groupby('Cliente_ID').agg(
    total_devolucoes=('Devolução_ID', 'count'),
    qtd_devolvida=('Quantidade', 'sum')
).reset_index()
```

```python
devolucoes_merge = devolucoes_merge.drop(columns=['Cliente_ID_x'])
```

```python
avaliacoes_merge = vendas.merge(
    avaliacoes[['Produto_ID', 'Avaliação']],
    on='Produto_ID',
    how='left'
)

avaliacoes_agg = avaliacoes_merge.groupby('Cliente_ID').agg(
    media_avaliacao=('Avaliação','mean')
).reset_index()
```

```python
tabela_final = (
    clientes[['Cliente_ID','Idade','Género','Cidade','Canal de Compra']]
    .merge(vendas_agg, on='Cliente_ID',how='left')
    .merge(devolucoes_agg, on='Cliente_ID', how='left')
    .merge(avaliacoes_agg, on='Cliente_ID',how='left'))
```
